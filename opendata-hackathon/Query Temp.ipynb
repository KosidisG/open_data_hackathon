```python
import requests
import xml.etree.ElementTree as ET

r = requests.get("http://weather.cyi.ac.cy/data/met/CyDoM.xml")


xmparsed = ET.fromstring(r.text)
```

```python
for i in xmparsed.findall("observations"):
    print(i[0].text)
    print(i[1].text)
    print(i[2][0].text, "\t", i[2][1].text)
    print("------------")
```

```python
data = []
for k, i in enumerate(xmparsed.findall("observations")):
    row = []
    data.append({})
    
    data[k] = {"station_id" : i[0].text}
    data[k]["time"] = i[1].text
    
    for num, z in enumerate(i):
        if num>1:
            data[k][i[num][0].text] = i[num][1].text.split(" ")[0]
```
